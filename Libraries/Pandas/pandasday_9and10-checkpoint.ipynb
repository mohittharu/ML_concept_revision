# create a Series from Dictionaries
-A dict can be passed as input and if no index is specified , then the dict keys are taken in a sorted order to construct index.(rows)
- if index is passed , the values in data corresponding to the lavels in the index will be pulled out. 


```python
import numpy as np 
import pandas as pd 
```

```python


data_dict= {
    'a': 0,
    'b': 1,
    'c': 2
}
s_dict = pd.Series(data_dict, index = ['b', 'a', 'c'])
s_dict
```

#Pivot

```python
tips.pivot_table(index='sex',values = ['total_bill','tip'], aggfunc= 'mean', observed= True)   # aggfun default nahuda mean use garni> categpries wise nikalney 
```

```python
tips.pivot_table(index='sex',columns= ['day','time'],values = ['total_bill','tip'],
                 aggfunc= {'total_bill':'mean', 'tip': 'sum'},              # '\' pani use garna sakinxa new line ko lagi 
                 margins= True, observed= True)
```
