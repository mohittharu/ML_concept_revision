# Boxplot/whisker plot
* min,q1,q2,q3,max
* iqr=q3-q1
* lowerlimit<--iqr-->upperlimit
* lowerlmit = q1-1.5*iqr
* upperlimit =q3+1.5*iqr
* min = nearest to lower limit
* max= nearest to upper limit
* Outliers = outside the upper limit
* q1=n+1/4
* q2 = n+1/2
* q3 = 3(n+1)/4
* 1. What is a Boxplot?

A boxplot (or whisker plot) summarizes data distribution based on five-number summary:

Minimum

First Quartile (Q1, 25%)

Median (Q2, 50%)

Third Quartile (Q3, 75%)

Maximum

It also highlights outliers (points beyond whiskers).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
matches = [100,110,110,110,120,120,130,140,140,150,170,220]
fig,ax= plt.subplots(figsize=(10,6))
ax.boxplot(matches)
ax.set(title= "Boxplot of matches",
      xlabel='Matches',
      ylabel = 'Count')
plt.show()
```

```python
matches = [[23,100,110,110,110,120,120,130,140,140,150,170,220],
          [23,100,110,110,110,120,120,130,140,140,150,170,220],
          [23,100,110,110,110,120,120,130,140,140,150,170,220]]
team = ['Team A','Team B','Team C']
colors = ['red','blue','orange']
fig,ax= plt.subplots(figsize=(10,6))
a= ax.boxplot(matches,tick_labels = team,patch_artist = True)
#coloring each boxes to make it we should make it patch_artist true
for patch, color in zip(a['boxes'], colors):
    patch.set_facecolor(color)
ax.set(title= "Boxplot of matches",
      xlabel='Matches',
      ylabel = 'Count')
plt.show()
```

```python
matches = [[23,100,110,110,110,120,120,130,140,140,150,170,220],
          [23,100,110,110,110,120,120,130,140,140,150,170,220],
          [23,100,110,110,110,120,120,130,140,140,150,170,220]]
team = ['Team A','Team B','Team C']
colors = ['red','blue','orange']
fig,ax= plt.subplots(figsize=(10,6))
a= ax.boxplot(matches,tick_labels = team, label= team,patch_artist = True)
#coloring each boxes to make it we should make it patch_artist true
for i,patch in enumerate(a['boxes']):
    patch.set_facecolor(colors[i])
ax.set(title= "Boxplot of matches",
      xlabel='Matches',
      ylabel = 'Count')
plt.legend(title = 'Team')
plt.show()
```
