# Bar plot

```python
# dictionary 

revenue = [2, 3, 4]
year = ['2010', '2011', '2012']

fig, ax = plt.subplots(figsize =(10,5))
bars= ax.bar(year, revenue, label=['2010','2011','2012'], color= ['Red', 'Green', 'Yellow'])   # width , height
ax.set(title='Revenue vs Year',
      xlabel = 'Year',
      ylabel = 'Revenue')

ax.legend( loc='upper left', title = 'Yearwise revenue')


# add count above the bars 
for bar in bars:
    print(bar)
    height = bar.get_height()
    print(height)
    ax.annotate(f'{height}',xy=(bar.get_x() + bar.get_width()/2, height))
                
              # xy=(1, height))

    
plt.show()
```

```python
import matplotlib.pyplot as plt



# Create a figure
fig = plt.figure(figsize=(12,12))

# Add a subplot to a 2x2 grid at position 1 (top-left)
ax1 = fig.add_subplot(2, 2, 1) # (221) or (2,2,1)
# ax1.plot([1, 2, 3], [4, 5, 6])
ax1.bar(year, revenue, label=['2010','2011','2012'], color= ['Red', 'Green', 'Yellow'])   # width , height
# ax1.set_title('Subplot 1')
ax1.set(title='Revenue vs Year',
      xlabel = 'Year',
      ylabel = 'Revenue')

ax1.legend(title = 'Yearwise revenue')


income = [2000, 3000, 4000]
year = ['2010', '2011', '2012']

# Add another subplot to the same 2x2 grid at position 2 (top-right)
ax2 = fig.add_subplot(222) # Using the 3-digit shorthand
# ax2.plot([1, 2, 3], [6, 5, 4])
ax2.barh(year, income, label=['2010','2011','2012'], color= ['brown', 'blue', 'orange'])   # width , height
# ax2.set_title('Subplot 2')
ax2.set(title='My yearly income',
      xlabel = 'Year',
      ylabel = 'Revenue')

ax2.legend(title = 'Yearly Income')


plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```

```python
data = [10,10,10,10]

Labels= ['A', 'B', 'C', 'D']

explode_area=[0,0,0,0.1]    # selected value laii bahira nikalna use garinxa 
fig, ax= plt.subplots(figsize =(8,8))
ax.pie(data,labels=Labels , autopct= '%1.2f%%', colors= ['Red','green','blue','yellow'],startangle=0, 
      hatch=['**O', 'oO', 'O.O', '.||.'],   # hatch tarika ho dot dot lyauney
      explode= explode_area,
       shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9} )   # shadow change garna ko lagi use vako ho 



ax.set(title= 'My data',
      xlabel='data',
      ylabel='labels')
plt.legend()
plt.show()
```
