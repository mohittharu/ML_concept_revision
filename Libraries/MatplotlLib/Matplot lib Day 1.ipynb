```python
#prepare data
x= [1,2,3,4]
y=[10,5,30,30]
# setup plot
fig,ax = plt.subplots(figsize =(10,5))   # height,width

#plot data
ax.plot(x,y,color = "red", marker= 'o', ms = 20)   # x,y cordinate and we add colors 
ax.set(title ="Prince vs Count",
      xlabel="Prince",
      ylabel="Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Add a subplot to a 2x2 grid at position 1 (top-left)
ax1 = fig.add_subplot(2, 2, 1) # (221) or (2,2,1)
ax1.plot([1, 2, 3], [4, 5, 6])
# ax1.set_title('Subplot 1')

# Add another subplot to the same 2x2 grid at position 2 (top-right)
ax2 = fig.add_subplot(222) # Using the 3-digit shorthand
ax2.plot([1, 2, 3], [6, 5, 4])
# ax2.set_title('Subplot 2')

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```
