```python
# mean

from scipy import stats

speed = [99, 22, 66, 77, 88, 87 , 93, 99, 78, 82, 97]

x = stats.mean(speed)
print(x)
```

```python
# median 
import numpy as np 

x = np.median(speed)
print(x)
```

```python
from scipy import stats

x= stats.mode(speed)
print(x)
```
