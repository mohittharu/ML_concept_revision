# Histogram Plot

### -A histogram is a type of bar plot that visualizes the distribution of numerical data by grouping it into ranges,
### -called bins.
### -Each bar in a histogram represents the frequency or number of data points that fall within a specific bin.
### -Histograms are useful for understanding the shape, spread, and central tendency of a dataset

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import colors
from matplotlib.ticker import PercentFormatter

# Create a random number generator with a fixed seed for reproducibility
rng = np.random.default_rng(19680801)
```

```python
rng
```

```python
N_points = 100000
n_bins = 20

# Generate two normal distributions
dist1 = rng.standard_normal(N_points)
dist2 = 0.4 * rng.standard_normal(N_points) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(dist1, bins=n_bins)
axs[1].hist(dist2, bins=n_bins)

plt.show()
```

# another way of ploting 

```python
# creating array for the plotting 
x=[1,2,3,4]
y=['A','B','C','D']

fig, ax= plt.subplots(ncols=2,nrows=2, figsize=(10,10))


ax[0,0].scatter(x,y, color="blue", marker='*', s=50)
ax[0,0].set_title('Scatter')
ax[0,0].legend()

ax[0,1].bar(x,y, label=['A','B','C','D'], color= ['Red', 'Green', 'Yellow'])
ax[0,1].set_title('Bargraph')
ax[0,1].legend(title='My detail')

ax[1,0].pie(x,labels=y , autopct= '%1.2f%%', colors= ['Red','green','blue','yellow'])
ax[1,0].set_title('Pie')
ax[1,0].legend()

             
ax[1,1].plot(x,y, color="blue", marker='*')
ax[1,1].set_title('Line')
ax[1,1].legend()


             
plt.show()
```
