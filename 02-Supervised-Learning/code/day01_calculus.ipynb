Derivatives help us understand the rate of change of a function,and we use them in gradient descentto optimize model parameters.

```python
import numpy as np 
import matplotlib.pyplot as plt 
def f(x):
    return x**2

def f_prime(x):
    return 2*x

x = np.linspace(-10, 10, 400)
y= f(x)
y_prime = f_prime(x)


plt.figure(figsize=(10,6))
plt.plot(x, y, label = '$f(X) = x^2$' , color= 'blue')
plt.plot(x, y_prime , label ='$f(X) = x^2$', color = 'red' , linestyle = '--')
plt.title("Function and its Derivatives")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gradient Descent function 
def gradient_descent(learning_rate, num_iterations):
    x_current = 10 # starting point 
    x_history = [x_current]

    for _ in range(num_iterations):
        grad = f_prime(x_current) #Gradient (derivatives)
        x_current= x_current - learning_rate * grad  #update step 
        x_history.append(x_current)

    return x_history

#Parameters 

learning_rate = 0.1
num_iterations = 20

#Run Gradient Descent 
x_history = gradient_descent(learning_rate, num_iterations)

# Plot the results 
x = np.linspace(-10, 10, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = x^2$', color='blue')
plt.scatter(x_history, f(np.array(x_history)), color='red', label="Gradient Descent Steps", zorder=5)
plt.title("Gradient Descent to Minimize $f(x) = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Integrals give us the area under a curve, and in ML, we often use them to compute areas under probability distributuions or total loss.

```python
from scipy.integrate import quad 
# Define the function for integration (eg. x^2)
def f_integral(x):
    return x**2

# Integrate the function from 0 to 3
integral_value, error = quad(f_integral, 0, 3)

# print the result 
print (f"Integral of x^2 from 0 to 3 : {integral_value}")
```

```python

# Plotting the function and area under the curve
x = np.linspace(0, 3, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = x^2$', color='blue')
plt.fill_between(x, y, where=(x <= 3), color='skyblue', alpha=0.5, label='Area under $f(x)$')
plt.title("Area Under the Curve of $f(x) = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 400)
y1 = np.sqrt(1 - x**2)
y2 = -y1

def transformation(x, y):
    return 9*x + 4*y, 4*x + 3*y

# original circle
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'b')
plt.axis('equal')
plt.title("Unit Circle")
plt.grid(True)
plt.show()

# transformed curve
x_new1, y_new1 = transformation(x, y1)
x_new2, y_new2 = transformation(x, y2)

plt.figure(figsize=(6, 6))
plt.plot(x_new1, y_new1, 'r')
plt.plot(x_new2, y_new2, 'r')
plt.axis('equal')
plt.title("Transformed Circle (Ellipse)")
plt.grid(True)
plt.show()
```
