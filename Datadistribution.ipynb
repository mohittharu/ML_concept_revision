```python
# histogram
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 250)

plt.hist(x, 5)
plt.show()
```

```python
# For big data
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100)
plt.show()
```

```python
# Normal data distribution -> an array where the values are concentrated around a given value.

import numpy 
import matplotlib.pyplot as plt

x= numpy.random.normal(5.0, 1.0, 10000)

plt.hist(x, 100)
plt.show()
```

```python
# Scatter Plot 

import numpy 
import matplotlib.pyplot as plt 

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]

y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x,y)
plt.show()
```

```python
# random data distribution using scatter 
#->two arrays that are both filled with 1000 random numbers from a normal data distribution.
import numpy as np 
import matplotlib.pyplot as plt 

x= numpy.random.normal(5.0, 1.0, 1000)
y= numpy.random.normal(10.0, 1.0, 1000)

plt.scatter(x, y)
plt.show()
```
