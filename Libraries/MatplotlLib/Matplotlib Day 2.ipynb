```python
x= [1,2,3,4]
```

```python
for i in x:
    print(np.sin(1))
```

```python
# PLOT THE LEGEND VALUE IN THE FIGURE 

fig,ax = plt.subplots(figsize = (15,5))
ax.scatter(a,np.sin(a), color= 'red', marker='*', s=50)
ax.scatter(a,np.cos(a), color= 'blue', marker='o', s=100)

ax.set(title='Scatter Sinwave Plot',
      xlabel=' X-axis',
      ylabel= 'Y-axis')
plt.legend(['Sin Wave', 'Cos Wave'], loc='upper right', title = 'Legend for Sin and Cos')
plt.show()
```

```python
# another way to plot the legend value  IN AXIS


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()


ax.plot(x, np.sin(x), '--b', label='Sine', )
ax.plot(x, np.cos(x), c='r', label='Cosine', )
ax.axis('equal')

leg = ax.legend(loc="lower left")
```

```python

# ANOTHER WAY TO PLOT THE LEGEND VALUE BY PLOT FUNCTION  

import numpy as np
import matplotlib.pyplot as plt

# X-axis values
x = np.arange(5)

# Y-axis values
y1 = [1, 2, 3, 4, 5]

# Y-axis values
y2 = [1, 4, 9, 16, 25]

# Function to plot
plt.plot(x, y1, label='Numbers')
plt.plot(x, y2, label='Square of numbers')

# Function add a legend
plt.legend(loc='upper right')

# function to show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Correct slicing
x1 = np.array(df['Price'].iloc[0:2])
y1 = np.array(df['Doors'].iloc[0:2])

x2 = np.array(df['Price'].iloc[2:5])
y2 = np.array(df['Doors'].iloc[2:5])

# Plotting
plt.scatter(x1, y1, color='blue', label='Group 1')
plt.scatter(x2, y2, color='red', label='Group 2')

# Correct function calls for labels and title
plt.xlabel= 'Price'
plt.ylabel= 'Doors'
plt.title= 'Price of Doors'
plt.legend()

plt.show()
```
